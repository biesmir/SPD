Generowanie losowych danych dla wariantu 2 maszynowego:

```python
harmonogram = Schedule([])
harmonogram.make_random(7,2)
print(harmonogram)
```

Test algorytmu Johnsona dla wariantu dwumaszynowego i losowych danych:

Obliczanie $C_{max}$ dla wszystkich permutacji (wartość $C_{max}$ odnosi się do kolejności pod nią):

```python
def permute(xs, low=0):
    if low + 1 >= len(xs):
        yield xs
    else:
        for p in permute(xs, low + 1):
            yield p
        for i in range(low + 1, len(xs)):
            xs[low], xs[i] = xs[i], xs[low]
            for p in permute(xs, low + 1):
                yield p
            xs[low], xs[i] = xs[i], xs[low]

for i in permute(harmonogram.joblist):
    print(harmonogram)
    print("cmax = " + str(harmonogram.cmax()))
    
```
