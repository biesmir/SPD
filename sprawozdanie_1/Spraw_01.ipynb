Obliczanie $C_{max}$ dla wszystkich permutacji (wartość $C_{max}$ odnosi się do kolejności pod nią):

```python
def permute(xs, low=0):
    if low + 1 >= len(xs):
        yield xs
    else:
        for p in permute(xs, low + 1):
            yield p
        for i in range(low + 1, len(xs)):
            xs[low], xs[i] = xs[i], xs[low]
            for p in permute(xs, low + 1):
                yield p
            xs[low], xs[i] = xs[i], xs[low]

for i in permute(harmonogram.joblist):
    print("cmax = " + str(harmonogram.cmax()))
    print(harmonogram)
    
    
```

Sprawdzenie poprawności poprzez stworzenie losowych serii danych testowych i sporządzenie tabeli z wynikami:

```python
from random import randint
import plotly.plotly as py
import plotly.figure_factory as ff
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
from plotly.graph_objs import *
init_notebook_mode()



mcmax_list = []
mkcmax_list = []
jhcmax_list = []
jobs = []
machines = []


for i in range(10):
    harmonogram = Schedule([])
    harmonogram.make_random(randint(3,5), 2)
    mkcmax = 0
    mcmax = harmonogram.cmax()
    
    jobs.append(harmonogram.number_of_jobs)
    machines.append(harmonogram.number_of_machines)
    
    harmonogram.johnson()
    jhcmax_list.append(harmonogram.cmax())
    
    for j in permute(harmonogram.joblist):
        if harmonogram.cmax() < mcmax:
            mcmax = harmonogram.cmax()
        if harmonogram.cmax() > mkcmax:
            mkcmax = harmonogram.cmax()
        
    
    mcmax_list.append(mcmax)
    mkcmax_list.append(mkcmax)

for i in range(10):
    harmonogram = Schedule([])
    harmonogram.make_random(randint(3,5), 3)
    mkcmax = 0
    mcmax = harmonogram.cmax()
    
    jobs.append(harmonogram.number_of_jobs)
    machines.append(harmonogram.number_of_machines)
    
    harmonogram.johnson()
    jhcmax_list.append(harmonogram.cmax())
    
    for j in permute(harmonogram.joblist):
        if harmonogram.cmax() < mcmax:
            mcmax = harmonogram.cmax()
        if harmonogram.cmax() > mkcmax:
            mkcmax = harmonogram.cmax()
        
    
    mcmax_list.append(mcmax)
    mkcmax_list.append(mkcmax)
            
trace = go.Table(
    header=dict(values=['Ilość maszyn', 'Ilość zadań', 'największe cmax', 'najmniejsze cmax', 'cmax dla algorytmu Johnsona']),
    cells=dict(values=[machines,
                        jobs,
                        mkcmax_list, 
                        mcmax_list,
                        jhcmax_list]))
    
layout = Layout(
    showlegend=False,
    height=600,
    width=600,
)

data = [trace] 
fig = dict( data=data, layout=layout )
iplot(fig) 
```
